```python
import pandas as pd
import numpy as np
```

```python
import nltk
nltk.download("punkt")
nltk.download("stopwords")
from nltk import word_tokenize
from nltk.corpus import stopwords
english_stop_words = stopwords.words('english')
```

```python
def removeStopWords(text):
    tokenized_word = word_tokenize(text)
    filtered_word = " ".join([word for word in tokenized_word if word not in english_stop_words])
    return filtered_word
stop_words_removed = [removeStopWords(sentence) for sentence in benefitsReviewDataset["benefitsSentimentProcessed"]]
```

```python
benefitsReviewDataset["benefitsSentimentProcessed"] = stop_words_removed 
```

```python
benefitsReviewDataset
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
nltk.download("averaged_perceptron_tagger")
from nltk.corpus import wordnet

lemmatizer = WordNetLemmatizer()
```

```python
def nltkTagToWordnetTag(NLTKtag):
    if NLTKtag.startswith('J'):
        return wordnet.ADJ
    elif NLTKtag.startswith('V'):
        return wordnet.VERB
    elif NLTKtag.startswith('N'):
        return wordnet.NOUN
    elif NLTKtag.startswith('R'):
        return wordnet.ADV
    else:
        return None;   
```

```python
def lemmatizeText(text):
    nltkTag = nltk.pos_tag(word_tokenize(text))
    #returns tuple( token, word_net )
    wordNetTag = map(lambda item: (item[0], nltkTagToWordnetTag(item[1])), nltkTag )
    lemmatizedText = []
    for word, tag in wordNetTag:
        if tag is None:
            lemmatizedText.append(word)
        else:
            lemmatizedText.append(lemmatizer.lemmatize(word,tag))
    return " ".join(lemmatizedText)
```

```python
lemmatizeText("harris mentioned that he would have cried") #example of lemmatization
```

```python
lemmatizedReviews = benefitsReviewDataset["benefitsSentimentProcessed"].apply(lambda text: lemmatizeText(text))
```

```python
benefitsReviewDataset["benefitsSentimentProcessed"] = lemmatizedReviews
benefitsReviewDataset.loc[0,:] #difference
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white")
from nltk import FreqDist
```

```python
allWordsString = " ".join([text for text in benefitsReviewDataset["benefitsSentimentProcessed"]])
allWords = allWordsString.split()
wordDist = FreqDist(allWords)
wordDist
```

```python
wordsFrequency = pd.DataFrame({"word": list(wordDist.keys()), "count" : list(wordDist.values()) })
#top 10 occuring words
wordsFrequencySubsetedAndSorted = wordsFrequency.nlargest(columns="count", n = 10)
wordsFrequencySubsetedAndSorted.sort_values('count', inplace=True)
```

```python
wordsFrequencySubsetedAndSorted
```

```python
plt.figure(figsize=(10,10))
axis = plt.barh(wordsFrequencySubsetedAndSorted["word"], width= wordsFrequencySubsetedAndSorted["count"])
axis.ylabel = "count"
plt.show()
```
